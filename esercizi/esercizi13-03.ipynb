```python
lower_limit, upper_limit = np.percentile(a =df, q =[0.5,99.5])
print(lower_limit)
print(upper_limit)
```

```python
df_filtered = df[(df['X']> lower_limit) & (df['X'] < upper_limit) & 
                 (df['Y']> lower_limit) & (df['Y'] < upper_limit) & 
                 (df['Z']> lower_limit) & (df['Z'] < upper_limit)]
print(df_filtered)
```

```python
df_filtered2 = df[(df> lower_limit) & (df < upper_limit)]
print(df_filtered2)
```
